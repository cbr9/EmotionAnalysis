```python
import pandas as pd
import json
import csv
```

```python
with open("unified-dataset.json") as json_file:
    data = []
    for line in json_file.readlines():
        data.append(json.loads(line)) 
df = pd.json_normalize(data)
```

```python
headers = df.columns
headers = [header.split(".")[1] if "." in header else header for header in headers ]
df.columns = headers
df
```

```python
tec = df[df["source"] == "tec"]
tec.dropna(axis=1, inplace=True)
labels = ["joy", "anger", "sadness", "disgust", "fear", "surprise"]
tec["label"] = tec[labels].idxmax(1)
tec.drop(columns=labels + ["labeled", "annotation_procedure", "emotion_model", "id", "domain", "source"], inplace=True)
# tec.to_csv("tec.tsv", sep="\t", encoding="utf8", index=False)
tec["n_tokens"] = tec["text"].str.split(" ").map(len)
print(tec["label"].value_counts())
tec.groupby("label")["n_tokens"].mean()
```

```python
labels = ["joy", "anger", "sadness", "disgust", "fear", "trust", "surprise", "noemo", "confusion", "anticipation"]
electoral_tweets = df[df["source"] == "electoraltweets"]
electoral_tweets.dropna(axis=1, inplace=True)
electoral_tweets["label"] = electoral_tweets[labels].idxmax(1)
electoral_tweets.drop(columns=labels + ["emotion_model", "domain", "labeled", "source", "id"], inplace=True)
# electoral_tweets.to_csv("electoral_tweets.tsv", sep="\t", encoding="utf8", index=False)

electoral_tweets["n_tokens"] = electoral_tweets["text"].str.split(" ").map(len)
print(electoral_tweets["label"].value_counts())
electoral_tweets.groupby("label")["n_tokens"].agg(["mean", "std"])
```
